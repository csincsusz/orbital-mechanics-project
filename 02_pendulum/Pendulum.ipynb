```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(y, dt):
    dalpha = y[1]*dt
    domega = -g*np.sin(y[0])/l * dt
    return np.array([dalpha, domega])
```

```python
def Euler(y):
    return y + f(y, dt)
```

```python
def E(y):
    E_kin = 1/2 * theta * y[1]**2
    E_pot = m * g * l * (1-np.cos(y[0]))
    return E_kin + E_pot
```

```python
l = 1.0
m = 1.0
theta = m * l**2
g = 9.81
alpha = 0.7
omega = 0.0
t_max = 10.0
dt = 0.01
```

```python
y = np.array([alpha, omega])
Y = [y]
Alpha = [alpha]
Omega = [omega]
t = 0.0
T = [t]
Energy = [E(y)]
```

```python
while t <= t_max:
    y = Euler(y)
    Y.append(y)
    Alpha.append(y[0])
    Omega.append(y[1])
    t += dt
    T.append(t)
    Energy.append(E(y))
```

```python
#plt.plot(T, Alpha)
#plt.plot(T, Omega)
plt.plot(T, Energy)
```
