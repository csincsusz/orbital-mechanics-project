```python
def newton(x, n, f, f_dot):
    return x - f(x, n)/f_dot(x)
```

```python
def f(x, n):
    return x**2 - n
```

```python
def f_dot(x):
    return 2*x
```

```python
n = 10
x = n/2
for i in range(5):
    x = newton(x, n, f, f_dot)
```

```python
print(x)
```
