```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(y, dt):
    v = y[1]
    dh = v*dt
    dv = (-g -(C*v*abs(v)) / m) * dt
    return np.array([dh, dv])
```

```python
def Euler(y):
    return y + f(y, dt)
```

```python
g = 9.81
C = 0.1
m = 1

t = 0.0
dt = 0.01
y = np.array([1000, 0])
T = [t]
Y = [y]
H = [y[0]]
V = [y[1]]

t_predict = 2*y[1]/g
```

```python
while y[0] >= 0:
    y = Euler(y)
    T.append(T[-1] + dt)
    Y.append(y)
    H.append(y[0])
    V.append(y[1])
```

```python
dt_prime = dt * H[-2] / (H[-2] - H[-1])
y_corrected = np.array([H[-2], V[-2]])
y_corrected = y_corrected + f(y_corrected, dt_prime)

H[-1] = y_corrected[0]
V[-1] = y_corrected[1]
T[-1] = T[-2] + dt_prime
```
